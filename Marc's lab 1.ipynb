```python
from IPython.display import IFrame, Image, display, Latex, Math
import pandas as pd
import numpy as np
```

```python
#calling in datatset from google spreadsheet
url = "https://docs.google.com/spreadsheets/d/1LkJbJGbmQ5MVtCc8oG_vFXalrMDDS44lDSz87GiUloQ/pub?output=csv"
vehicle = pd.read_csv(url)
vehicle.head()
vehicle
```

```python
#delete columns that are not needed
vehicleDrop = vehicle.drop(['cyl','disp','drat','wt','qsec','vs','am','gear','carb'], axis=1)
vehicleDrop
```

```python
#tidy up the data set combining mpg and hp to the same collumn with their corresponding values
#in the value column. 
tidy_vehicle = pd.melt(vehicleDrop, id_vars=["Model"],
                      var_name="mpg,hp", value_name="value")
tidy_vehicle.head()
tidy_vehicle
```
